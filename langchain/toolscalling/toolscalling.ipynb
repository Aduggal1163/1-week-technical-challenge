Tool Binding : a process where u connect ur llm with a given tool


```python
import os
from dotenv import load_dotenv
load_dotenv()
GOOGLE_API_KEY = os.getenv("GOOGLE_API_KEY")
os.environ["GOOGLE_API_KEY"]=GOOGLE_API_KEY
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.tools import tool
from langchain_core.messages import HumanMessage
import requests
```

```python
#Tool create
@tool #tool decorator
def multiply(a: int , b : int) -> int:
    """Given 2 numbers a and b this tool returns their product"""
    return a*b
```

```python
print(multiply.invoke({"a":3,"b":5}))
print(multiply.name)
print(multiply.description)
print(multiply.args)
```

```python
#tool binding
llm = ChatGoogleGenerativeAI(model="gemini-3-large", temperature=0)
```

```python
llm_with_tools=llm.bind_tools([multiply]) # yeh meine tools ko bind krdiya agr ik se jyada ho toh , daalke hojega
```
