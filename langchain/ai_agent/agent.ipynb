```python
import os
from langchain_google_genai import ChatGoogleGenerativeAI
from dotenv import load_dotenv
load_dotenv()
GOOGLE_API_KEY = os.getenv("GOOGLE_API_KEY")
os.environ["GOOGLE_API_KEY"] = GOOGLE_API_KEY
```

```python
from langchain_community.tools import DuckDuckGoSearchRun
#Tool (Built-In)
search_tool = DuckDuckGoSearchRun()
# results = search_tool.invoke("MS Dhoni")
```

```python
import requests
@tool
def get_weather_data (city : str) -> str:
    """
        this function fetches the current weather data for a given city
    """
    url = f'https://api.weatherstack.com/current?access_key=0078760375d955ed5fa7d9ac3ed85c88&query={city}'
    response = requests.get(url)
    return response.json()
```

```python
# Initialize LLM (Gemini Model)
# ==============================
# temperature=0 â†’ makes output deterministic (less randomness)
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=0)
```

```python
llm.invoke("Hi")
```

AGENT CREATE KARNA SHURU

```python
from langgraph.prebuilt import create_react_agent
```

```python
# ==============================
# Create ReAct Agent
# ==============================
# ReAct = Reason + Act
# The agent can:
#   - Think (reason about what to do)
#   - Use tools (search_tool)
#   - Observe tool result
#   - Give final answer #pulls standard ReAct agent prompt react means reasoning+action
agent = create_react_agent(
    model=llm,
    tools=[search_tool,get_weather_data],
    # debug=True
)
```

```python
# Invoke Agent
# ==============================
# Messages format required in LangGraph:
# List of (role, content) tuples
response = agent.invoke({
    "messages": [
        ("user", "Find the capital of Madhya Pradesh, then find its current weather condition")
    ]
})
```

```python
# response['messages'][-1].content[0]['text']
response['output']

# ðŸ”Ž Why This Works
# response['messages'] â†’ list of messages
# [-1] â†’ last message
# .content â†’ list of content blocks
# [0] â†’ first block
# ['text'] â†’ actual text string
```
