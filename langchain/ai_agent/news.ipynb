“Find the latest news about Tesla, summarize it, and give a sentiment score (positive/negative/neutral).”

```python
import os
import requests
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.tools import tool
from dotenv import load_dotenv
load_dotenv()
GOOGLE_API_KEY = os.getenv("GOOGLE_API_KEY")
os.environ["GOOGLE_API_KEY"] = GOOGLE_API_KEY
```

```python
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=0)
```

```python
from langchain_community.tools import DuckDuckGoSearchRun
search_result =DuckDuckGoSearchRun()
# search_result.run("MS DHONI")
```

```python
@tool
def summarize_news(news: str)->str:
    """
     this functions summarize the entire news 
    """
    return llm.invoke(f"summarize this entire topic in 1-2 lines {news}").content
```

```python
@tool
def sentiment_score(text : str)->str:
    """
    this functions tells  sentiment of the text: Positive, Negative, Neutral
    """
    return llm.invoke(f"Analyze the sentiment of this text and answer only with Positive, Negative, or Neutral:{text}").content
```

```python
from langgraph.prebuilt import create_react_agent

agent = create_react_agent(
    model=llm,
    tools = [search_result,summarize_news,sentiment_score],
    debug=True
)
```

```python
response = agent.invoke (
    {
       "messages":[ ('user',"Find the latest news about tesla and then summarize it in 1-2 lines and also tell sentiment whether it is a Positive, Negative, Neutral")]
    }
)
```
